## Introduction

This project analyzes property price data using statistical techniques.
The goal of this analysis is to understand the distribution of property
prices and identify the relationship between the number of rooms and
the property price.

Various statistical measures such as mean, median, mode, standard
deviation, and variance are used along with visualization techniques
to better understand the dataset.

## Dataset Loading

In this step, the property dataset is loaded into the Python
environment using the pandas library. The dataset contains
information about property prices, number of rooms, location,
and other features.

import pandas as pd

df = pd.read_csv("property.csv")
df.head()

## Dataset Information

The dataset information provides details about the number
of rows, columns, and data types present in the dataset.
This helps in understanding the structure of the dataset.

```python
df.info()
```

## Descriptive Statistics

Descriptive statistics help summarize the main characteristics
of the dataset. It provides important measures such as count,
mean, standard deviation, minimum, and maximum values.

```python
df.describe()
```

```python
df.columns
```

## Central Tendency Measures

Central tendency measures such as mean, median, and mode
are used to understand the typical value of property prices
in the dataset.

```python
df["Price"].mean()
```

```python
df["Price"].median()
```

```python
df["Price"].mode()
```

## Variability Measures

Standard deviation and variance are used to measure the
spread or variability of property prices in the dataset.

```python
df["Price"].std()
```

```python
df["Price"].var()
```

## Histogram Visualization

A histogram is used to visualize the distribution of property
prices in the dataset. It helps to understand how prices are
spread across different ranges.

```python
import matplotlib.pyplot as plt

df["Price"].hist()
plt.show()
```

## Rooms vs Price Analysis

This analysis examines how the number of rooms in a property
affects its price. Generally, properties with more rooms tend
to have higher prices.

```python
df.groupby("Rooms")["Price"].mean()
```

## Scatter Plot Visualization

A scatter plot is used to visualize the relationship between
the number of rooms and property prices.

```python
df.plot.scatter(x="Rooms", y="Price")
plt.show()
```

## Hypothesis Testing

A one-sample t-test is performed to determine whether the
average property price is significantly different from
1,000,000.

```python
from scipy import stats

stats.ttest_1samp(df["Price"], 1000000)
```

## Conclusion

The analysis shows that the average property price in the
dataset is approximately 1,075,684.

Properties with more rooms generally tend to have higher
prices. The statistical test confirms that the average
property price is significantly different from 1,000,000.
